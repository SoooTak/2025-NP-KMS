1. 두 개의 매개변수 n, m을 전달받아 m x n개의 * 상자를 출력하는 프로그램을 함수로 작성

예: 2, 4 → ****
            ****

```python
def stars(n, m):
    for i in range(n):
        print('*' * m)
stars(2, 4)
```

2. 하나의 숫자를 전달받아 숫자의 자리 합을 구하는 함수를 작성
예: 123 → 1+2+3 = 6

```python
def sum_of_digits(x):
    x = abs(int(x))
    s = 0
    while x > 0:
        s += x % 10
        x //= 10
    return s

print(sum_of_digits(123))  # 6
```

3. 두 개의 문자열이 서로 다른 처음 위치를 반환하는 함수를 작성. 두 개의 문자열이 같으면 -1을반환

```python
def first_diff_pos(a, b):
    n = min(len(a), len(b))
    for i in range(n):
        if a[i] != b[i]:
            return i
    if len(a) != len(b):
        return n
    return -1

print(first_diff_pos("hello", "heLlo"))  # 2
print(first_diff_pos("abc", "abc"))      # -1
print(first_diff_pos("abc", "ab"))       # 2
```

4. 숫자를 전달받아 그 수의 약수를 리스트로 반환하는 함수를 작성

```python
def divisors(n):
    n = int(n)
    if n <= 0:
        return []
    result = []
    for i in range(1, n + 1):
        if n % i == 0:
            result.append(i)
    return result

print(divisors(12))  # [1, 2, 3, 4, 6, 12]
```

5. 문자열과 하나의 문자를 전달받아 문자열에서 문자의 위치를 모두 찾아 리스트로 반환하는 함수를 작성

```python
def find_all_positions(s, ch):
    return [i for i, c in enumerate(s) if c == ch]

print(find_all_positions("banana", "a"))  # [1, 3, 5]
```

6. 재귀 함수를 이용하여 1부터 100까지의 합을 계산하는 프로그램

```python
def recursive_sum(n):
    if n <= 1:
        return max(n, 0)
    return n + recursive_sum(n - 1)

print(recursive_sum(100))  # 5050
```
